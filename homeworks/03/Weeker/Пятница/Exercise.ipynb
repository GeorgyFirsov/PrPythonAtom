# Occupation

### Introduction:


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('u.user', sep = '|')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occst = set(users.occupation)
data = []
for item in occst:
    ratio = (users.loc[(users['occupation'] == item)&(users['gender'] == 'M')])['user_id'].count()/(users.loc[(users['occupation'] == item)])['user_id'].count()
    data.append({'occupation': item, 'ratio': ratio})
result = pd.DataFrame(data).sort_values('ratio', ascending = False)
result
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
occst = list(set(users.occupation))
data = []
for item in occst:
    data.append({'occupation': item, 'min age': (users.loc[(users['occupation'] == item)])['age'].min(), 'max age': (users.loc[(users['occupation'] == item)])['age'].max()})
result = pd.DataFrame(data)
result
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['gender', 'occupation']).mean()[['age']]
```

### Step 8.  For each occupation present the percentage of women and men

```python
occst = set(users.occupation)
data = []
for item in occst:
    MR = (users.loc[(users['occupation'] == item)&(users['gender'] == 'M')])['user_id'].count()/(users.loc[(users['occupation'] == item)])['user_id'].count()
    FR = 1 - MR
    data.append({'occupation': item, 'Men percentage': MR * 100, 'Women percentage': FR * 100})
result = pd.DataFrame(data)
result
```
