```python
import random

class Entry: # пришлось и тут и внутри класса объявлять, ибо работает только так
        def __init__(self, key, value):
            self.key = key
            self.value = value
            self.next = []
        def get_key(self):
            return self.key
        def get_value(self):
            return self.value
        def __eq__(self, other):
            if (self.key == other.key):
                return True
            else:
                return False

class HashMap:
    
    class Entry:
        def __init__(self, key, value):
            self.key = key
            self.value = value
            self.next = []
        def get_key(self):
            return self.key
        def get_value(self):
            return self.value
        def __eq__(self, other):
            if (self.key == other.key):
                return True
            else:
                return False
    
    def __init__(self, bucket_num = 64):
        self.array = []
        for _ in range(0, bucket_num):
            self.array.append(Entry(None, None))

    def get(self, key, default_value = None):
        index = self._get_index(self._get_hash(key))
        if self.array[index] == Entry(None,None):
            return default_value
        if self.array[index].key == key:
            return self.array[index].value
        else:
            for i in range(0, len(self.array[index].next)):
                if self.array[index].next[i].key == key:
                    return self.array[index].next[i].value
        return default_value
                
    def put(self, key, value):
        index = self._get_index(self._get_hash(key))
        if (self.array[index].key == None):
            self.array[index] = Entry(key,value)
            return
        else:
            if self.array[index].get_key() == key:
                self.array[index].value = value
            else:
                for i in range(0, len(self.array[index].next)):
                    if self.array[index].next[i].key == key:
                        self.array[index].next[i].value = value
                self.array[index].next.append(Entry(key, value))

    def __len__(self):
        result = 0
        for i in range(0, len(self.array)):
            if (self.array[i].get_value() != None):
                result = result + 1 + len(self.array[i].next)
            i = i + 1
        return result

    def _get_hash(self, key):
        return hash(key) 
    
    def _get_index(self, hash_value):
        return ((hash_value * 3) % len(self.array))

    def values(self):
        i = -1
        while True:
            i += 1
            j = -1
            if(i == len(self.array)):
                raise StopIteration
            if(self.array[i].get_key() != None):
                if j == -1:
                    result = self.array[i].get_value()
                    yield result
                    while j < len(self.array[i].next):
                        j += 1
                        if j == len(self.array[i].next):
                            i += 1
                            continue
                        if self.array[i].next[j].get_key() != None:
                            result = self.array[i].next[j].get_value()
                            yield result
        
    def keys(self):
        i = -1
        while True:
            i += 1
            j = -1
            if(i == len(self.array)):
                raise StopIteration
            if(self.array[i].get_key() != None):
                if j == -1:
                    result = self.array[i].get_key()
                    yield result
                    while j < len(self.array[i].next):
                        j += 1
                        if j == len(self.array[i].next):
                            i += 1
                            continue
                        if self.array[i].next[j].get_key() != None:
                            result = self.array[i].next[j].get_key()
                            yield result

    def items(self):
        i = -1
        while True:
            i += 1
            j = -1
            if(i == len(self.array)):
                raise StopIteration
            if(self.array[i].get_key() != None):
                if j == -1:
                    result = (self.array[i].get_key(), self.array[i].get_value())
                    yield result
                    while j < len(self.array[i].next):
                        j += 1
                        if j == len(self.array[i].next):
                            i += 1
                            continue
                        if self.array[i].next[j].get_key() != None:
                            result = (self.array[i].next[j].get_key(), self.array[i].next[j].get_value())
                            yield result
        
    def __str__(self): # Вывод в виде HM! { key: k1; value: v1 } { key: k2; value: v2 } ... !HM (Lenght: n)
        result = "HM!"
        for i in range(0, len(self.array)):
            if(self.array[i].get_key() != None):
                index = self._get_index(self._get_hash(i))
                result = result + " { key: "
                result = result + str(self.array[i].get_key())
                result = result + "; value: "
                result = result + str(self.array[i].get_value())
                result = result + " }"
                for j in range(0, len(self.array[i].next)):
                    result = result + " { key: "
                    result = result + str(self.array[i].next[j].get_value())
                    result = result + "; value: "
                    result = result + str(self.array[i].next[j].get_key())
                    result = result + " }"
                    
        result = result + " !HM (Length: {})".format(str(len(self)))
        return result

def test_hashmap_04():
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ()), ({"s":"v"}, {"v":"s"})]
    for k, v in entries:
        entry = HashMap.Entry(k, v)
        assert entry.get_key() == k
        assert entry.get_value() == v
    print("Test 4 part 01 passed")
        
    for i in range(len(entries)):
        entry_one = HashMap.Entry(entries[i][0], entries[i][1])
        for _ in range(10):
            j = random.randint(0, len(entries)-1)
            p = random.randint(0, len(entries)-1)
            entry_two = HashMap.Entry(entries[j][0], entries[p][1])
            if j == i:
                assert entry_one == entry_two
            else:
                assert entry_one != entry_two
    print("Test 4 part 02 passed")
    
def test_hashmap_05():
    hashmap = HashMap(10)
    assert sum(isinstance(v, list) for k, v in hashmap.__dict__.items()) == 1
    print("Test 5 part 01 passed")
    
    inner_list_name = [k for k, v in hashmap.__dict__.items() if isinstance(v, list)][0]
    for i in range(10):
        assert len(HashMap(i).__dict__[inner_list_name]) == i
    print("Test 5 part 02 passed")
    
def test_hashmap_06():
    hashmap = HashMap(10)
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ())]
    for k, v in entries:
        hashmap.put(k, v)
    for k, v in entries:
        assert hashmap.get(k) == v

    print("Test 6 part 01 passed")
    
    assert hashmap.get("nexit", "default") == "default"
    
    print("Test 6 part 02 passed")

test_hashmap_04()
test_hashmap_05()
test_hashmap_06()

hm = HashMap()
for i in range(0, 15):
    hm.put(i, i**2 - 8 + i)
print(hm)
v = hm.keys()
try:
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
    print(next(v))
except:
    pass
```

```python
class Entry:
    def __init__(self, key, value = None):
        self.key = key
        self.value = None
        self.next = []
            
    def get_value(self):
        return self.key  # да, немного странно, но так как сет использует заглушки вместо значений, то
                         # метод возвращает ключ (как "значение")
    def __eq__(self, other):
        if (self.key == other.key):
            return True
        else:
            return False

class HashSet(HashMap):
    def __init__(self, bucket_num = 64):
        self.array = []
        for _ in range(0, bucket_num):
            self.array.append(Entry(None))

    def get(self, key, default_value = None):
        index = self._get_index(self._get_hash(key))
        if self.array[index].key == None:
            return default_value
        if self.array[index].key == key:
            return key
        else:
            for i in range(0, len(self.array[index].next)):
                if self.array[index].next[i].key == key:
                    return key
        return default_value
        

    def put(self, key, value = "ENGAGED"): # вместо значения ставим заглушку
        index = self._get_index(self._get_hash(key))
        if self.array[index].key == None:
            self.array[index] = Entry(key, value)
            return
        else:
            if self.array[index].key != key:
                for i in range(0, len(self.array[index].next)):
                    if self.array[index].next[i].key != key:
                        continue
                self.array[index].next.append(Entry(key, value))

    def __len__(self):
        result = 0
        for i in range(0, len(self.array)):
            if (self.get(i) != None):
                result = result + 1 + len(self.array[i].next)
            i = i + 1
        return result
    
    def __str__(self): # Вывод в виде HS! { v1 } { v2 } ... !HS (Lenght: n)
        result = "HS!"
        for i in range(0, len(self.array)):
            if(self.get(i) != None):
                result = result + " { "
                result = result + str(self.array[self._get_index(self._get_hash(i))].get_value())
                result = result + " }"
                for j in range(0, len(self.array[i].next)):
                    result = result + " { "
                    result = result + str(self.array[i].next[j].get_value())
                    result = result + " }"
        result = result + " !HS (Length: {})".format(str(len(self)))
        return result

    def values(self):
        i = -1
        while True:
            i += 1
            j = -1
            if(i == len(self.array)):
                raise StopIteration
            if(self.array[i].key != None):
                if j == -1:
                    result = self.array[i].get_value()
                    yield result
                    while j < len(self.array[i].next):
                        j += 1
                        if j == len(self.array[i].next):
                            i += 1
                            continue
                        if self.array[i].next[j].key != None:
                            result = self.array[i].next[j].get_value()
                            yield result
    
a = HashSet()
for i in range(25, 35):
    a.put(64 % (i-20))
print(a)
print(len(a))
gen = a.values()
try:
    print("Values:")
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
except StopIteration:
    pass
```
